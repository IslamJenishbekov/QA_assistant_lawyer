#Гражданский кодекс 1

```python
with open("/content/drive/MyDrive/QA/orig_data/Гражданский Кодекс КР Часть I.txt", "r") as f:
  text = f.read().split("\n \n")[3: -2]
  text = "\n".join(text).split("\n")
  articles, article = list(), list()
  for i in text:
    if "Статья" in i and len(article) > 0:
      articles.append(article)
      article = list()
      s = i.split()
      if len(s[1]) > 4 and "-" not in s[1]:
        s[1] = s[1][:3] + "-" + s[1][3:]
        i = " ".join(s)
    article.append(i)
  articles.append(article)
```

```python
metadata = {} #format {index : {"book" : book, "article_num" : article_num, "link" : link}}
chunks = []
index = 0
link = "https://cbd.minjust.gov.kg/4/edition/7214/ru"
book = "Гражданский Кодекс КР Часть I"

for article in articles:
  article_num = article[0].split()[1][:-1]
  chunk = ""
  for i in range(1, len(article)):
    sen = article[i].strip()
    if len(sen) < 2 or sen.startswith("Глава") or sen.startswith("(") and sen.endswith(")"):
      continue
    elif sen.startswith("См. также:") or sen.startswith("Параграф"):
      article[i+1] = ""
      continue
    elif sen.split()[0][-1] == "." and sen.split()[0][0].isdigit() and len(chunk) > 0:
      chunks.append(chunk.strip())
      metadata[index] = {"book" : book, "article_num" : article_num, "link" : link}
      index += 1
      chunk = ""
    chunk += sen + "\n"
  if len(chunk) > 0:
    chunks.append(chunk.strip())
    metadata[index] = {"book" : book, "article_num" : article_num, "link" : link}
    index += 1
```

```python
def get_type_sign(sub_str):
    sub_str = sub_str.strip()
    if sub_str[0].isdigit() and sub_str[-1] == ")":
        return "digit"
    elif sub_str[0].isalpha() and sub_str[-1] == ")":
        return "alpha"
    elif sub_str[0] == "-":
        return "line"
    else:
        return "empty"

print(get_type_sign("1)"))
```

```python
def break_long_for_more_colons(chunk_index):
  global index
  text = chunks[chunk_index].split("\n")
  new_chunks = []
  sens_with_colon = []
  hierarchy = []
  chunk = ""
  for i in range(len(text)):
    sen = text[i]
    if len(sen) == 0:
      continue
    elif sen[-1] == ":":
      type_colon = get_type_sign(text[i+1].split()[0])
      hierarchy.append(type_colon)
      if len(chunk) > 0 and chunk.strip()[-1] != ":":
        new_chunks.append(chunk.strip())
        chunk = ""
        for j in sens_with_colon:
          chunk += j

      if len(hierarchy) > 1 and hierarchy[-1] == hierarchy[-2]:
        hierarchy.pop()
        del sens_with_colon[-1]
        chunk = ""
        for j in sens_with_colon:
          chunk += j
      sens_with_colon.append(sen + "\n")
    elif len(sens_with_colon) > 0:
      type_colon = get_type_sign(sen.split()[0])
      if type_colon == "empty":
        sens_with_colon = []
        hierarchy = []
        new_chunks.append(chunk.strip())
        chunk = ""
      elif len(hierarchy) > 1 and type_colon == hierarchy[-2]:
        hierarchy.pop()
        sens_with_colon.pop()
        new_chunks.append(chunk.strip())
        chunk = ""
        for j in sens_with_colon:
          chunk += j

    if len(chunk + sen) > 2000:
      new_chunks.append(chunk.strip())
      chunk = ''
      for j in sens_with_colon:
        chunk += j
      chunk += sen
    else:
      chunk += sen + "\n"
  if len(chunk) > 0:
    new_chunks.append(chunk.strip())


  chunks[chunk_index] = new_chunks[0]
  for i in range(1, len(new_chunks)):
    metadata[index] = metadata[chunk_index]
    chunks.append(new_chunks[i])
    index += 1
```

```python
for i in chunks:
  if len(i) > 2000 and i.count(":") > 1:
    break_long_for_more_colons(chunks.index(i))
```
